```python
iteMS = []
print("WelcoMe to todo app")
print("Press A to add iteM")
print("Press D to delete iteM")
print("Press DA to delete all iteMS")
print("Press Q to quit")

def displayIteMSList():
    print()
    for iteM in iteMS:
        print(iteM)

def reMoveFroMList(paraM):
    totalIteMOccurence = iteMS.count(paraM)
    
    if totalIteMOccurence == 0:
        print("{0} doesnot exist in list".format(paraM))
    
    else:
        for i in range(totalIteMOccurence):
            iteMS.remove(paraM)

while True:
    userAction = input("Enter the action that you want to perforM ")
    
    if userAction.upper() == "A":
        iteMS.append(input("Enter iteM please "))
        displayIteMSList()
    
    elif userAction.upper() == "D":
        
        if len(iteMS) > 0:
            reMoveFroMList(input("Enter iteM to delete "))
            displayIteMSList()
            
        else:
            print("There is no iteM in list to delete")
    
    elif userAction.upper() == "DA":
        if len(iteMS) > 0:
            for i in range(len(iteMS)):
                iteMS.pop()
        print("The list is eMpty now")
        print(iteMS)
    
    elif userAction.upper() == 'Q':
        break
    
    else:
        print("Please input valid action")
```
