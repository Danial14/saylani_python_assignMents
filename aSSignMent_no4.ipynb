```python
country = 'country'
population = "population"
facts = 'facts'
cities = {
    "Karachi" : {country : "Pakistan",
                 population : 15741000,
                 facts : ['Biryani', "Mazar-e-Quaid", "Clifton beach"]
                },
    "Lahore" : {country : "Pakistan",
                population : 12188000,
                facts : ['Data Darbar', "Murgh cholay", 'Minar-e-Pakistan']
               },
    "IslaMabad" : {country : "Pakistan",
                   population : 1095064,
                   facts : ["Faisal Masjid", 'Capital of Pakistan', "Margala hillS"]
                  }
}

def displayCitiesInfo():
    for city, cityInfo in cities.items():
        print("Info about", city)
        for info in cityInfo.values():
            if type(info) == list:
                displayCityFacts(info)
            else:
                print(info)
        print()

def displayCityFacts(cityFacts):
    for fact in cityFacts:
        print(fact)
        
displayCitiesInfo()
```
